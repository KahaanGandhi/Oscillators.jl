Consider $\underline{three}$ identical masses coupled by four identical springs. Using Newton's second law, 
$$
\begin{aligned}
F_1 = -2kx_1 + kx_2 &\rightarrow m\ddot{x_1} + 2kx_1 - kx_2 = 0 \\
F_2 = kx_1 - 2kx_2 + kx_3 &\rightarrow m\ddot{x_2} + 2kx_2 - kx_1 - kx_3 = 0 \\
F_3 = kx_2 - 2kx_3 &\rightarrow m\ddot{x_3} + 2kx_3 - kx_2 = 0
\end{aligned}
$$
We make the ansatz $ x_j = A_j e^{i \omega t}$, where mass index $j = (1,2,3)$ since we have three atoms. Differentiating, 
$$
\begin{aligned}
x_j &= A_j e^{i \omega t} \\
\dot{x_j} &= i \omega A_j e^{i \omega t} \\
\ddot{x_j}&= -\omega^2 A_j e^{i \omega t}
\end{aligned}$$
The equations of motion become
$$
-\omega^2 \vec{x} = \omega_0^2 \begin{bmatrix} -2 & 1 & 0 \\ 1 &-2  &1 \\ 0 & 1 & -2 \end{bmatrix}\vec{x}, \text{ where } \omega_0 = \sqrt{\frac{k}{m}}.
$$
The eigenvalues of this matrix correspond to the square of normal mode frequencies. We find the eigenvalues and eigenvectors 
$$
\begin{aligned}
\omega_1 = \omega_0 \sqrt{2-\sqrt{2}} &\rightarrow e_1 = \begin{bmatrix} 1 \\ \sqrt{2} \\ 1 \end{bmatrix} \\
\omega_3 = \omega_0 \sqrt{2+\sqrt{2}} &\rightarrow e_2 = \begin{bmatrix} 1 \\ -\sqrt{2} \\ 1 \end{bmatrix} \\
\omega_2 = \omega_0 \sqrt{2} &\rightarrow e_3 = \begin{bmatrix} -1 \\ 0\\ 1 \end{bmatrix}
\end{aligned}
$$

Now, consider $\underline{four}$ masses coupled by 5 springs, once again holding each mass and spring constant. Using Newton's second law, 
$$
\begin{aligned}
F_1 &= -kx_1 + k(x_2 - x_1)  \\
F_2 &= -k(x_2 - x_1) + k(x_3 - x_2) \\
F_3 &= -k(x_3 - x_2) + k(x_4 - x_3) \\
F_4 &= -k(x_4 - x_3) -kx_4 
\end{aligned} \Longrightarrow
\begin{aligned}
m_1 \ddot{x_1} &= -2kx_1 + kx_2 \\
m_2 \ddot{x_2} &= -2kx_2 + kx_1 + kx_3 \\
m_3 \ddot{x_3} &= -2kx_3 + kx_2 + kx_4 \\
m_4 \ddot{x_4} &= -2kx_4 + kx_3
\end{aligned} \Longrightarrow
\begin{aligned}
\ddot{x_1} &= \omega_0^2 (-2x_1 + 1x_2)  \\
\ddot{x_2} &= \omega_0^2 (-2x_2 + 1x_1 + 1x_3)  \\
\ddot{x_3} &= \omega_0^2 (-2x_3 + 1x_2 + 1x_4)  \\
\ddot{x_4} &= \omega_0^2 (-2x_4 + 1x_3) 
\end{aligned} 
$$
Converting the system of differential equations into matrix form, 
$$
\begin{bmatrix}
\ddot{x_1} \\ \ddot{x_2} \\ \ddot{x_3} \\ \ddot{x_4}
\end{bmatrix} = \omega_0^2 \begin{bmatrix}
-2 & 1 & 0 & 0 \\
1 & -2 & 1 & 0 \\
0 & 1 & -2 & 1 \\
0 & 0 & 1 & -2
\end{bmatrix} \begin{bmatrix}
x_1 \\ x_2 \\ x_3 \\ x_4
\end{bmatrix}
$$
We make the ansatz $ x_j = A_j e^{i \omega t}$, where mass index $j = (1,2,3,4)$ since we have four masses. Differentiating, 
$$
\begin{aligned}
x_j &= A_j e^{i \omega t} \\
\dot{x_j} &= i \omega A_j e^{i \omega t} \\
\ddot{x_j}&= -\omega^2 A_j e^{i \omega t}
\end{aligned}
$$
Substiutitng back into the matrix equation and cancelling common terms, 
$$
-\omega^2 \vec{x} = \omega_0^2 \begin{bmatrix} 
-2 & 1 & 0 & 0 \\
1 & -2 & 1 & 0 \\
0 & 1 & -2 & 1 \\
0 & 0 & 1 & -2 \end{bmatrix}\vec{x}
$$

We can generalize a tridiagonal matrix for $\underline{N}$ identical masses coupled by N+1 identical springs:
$$
A_{N \times N} = \begin{bmatrix}
-2 & 1 & 0 & \cdots & 0 \\
1 & -2 & 1 & \cdots & 0 \\
0 & 1 & -2 & \ddots & \vdots \\
\vdots & \vdots & \ddots & \ddots & 1 \\
0 & 0 & \cdots & 1 & -2
\end{bmatrix}, \text{where} -\omega^2 \vec{x} = \omega_0^2 A \vec{x}
$$
We can then solve for the N eigenvalues and N corresponding eigenvectors of A, as shown in `N-CoupledChain.jl`.
