Slightly altering the setup of `3-BeadedString.ipynb`, we now have a string of mass M and lenght L, and we consider the limit $N \rightarrow \infty$ (and $a \rightarrow 0$). 

$$
\begin{align*}
& \quad \quad \quad \quad \quad y_1 \quad\quad \; y_2 \quad \quad \quad \quad \quad \quad y_N \\
& \quad\quad\quad\quad \;\;\; \uparrow \quad \;\;\; \quad\uparrow \quad \quad \quad \quad \; \quad \quad \uparrow \\
& \boxed{\phantom{aaaa}} == m_1  ==  m_2  ==  \cdots == m_N == \boxed{\phantom{aaaa}} \\
& \quad \quad \;\;\;  <------ L ------>
\end{align*}
$$

For this example, we use the following initial conditions:
$$ 
   \dot{y_0} = 0 \text{ , and }y_0 (x) = \begin{cases} 
      B & \text{if } \frac{L}{2} < x < \frac{3L}{4} \\
      0 & \text{else} \\
   \end{cases}
$$
For the beaded string, $\omega_n = 2 \sqrt{\frac{T}{ma}}sin\left(\frac{n \pi a}{2L}\right) = 2 \sqrt{\frac{T}{ma}}\sin\left(\frac{n \pi}{2(N+1)}\right)$. As $N \rightarrow \infty$, the argument of the sin becomes small enough to apply the small angle approximation $\sin(x) \approx x$. Defining mass per unit lenght $\mu \equiv \frac{m}{a}$, the normal mode frequency for a continous string therefore takes the form
$$
\omega_n = 2 \sqrt{\frac{T}{ma}}\left(\frac{n \pi a}{2L}\right) = \frac{n \pi}{L}\sqrt{\frac{T}{\mu}} = \frac{n \pi}{L}\sqrt{\frac{TL}{M}}
$$
For the beaded string, the eigenvector for the $n^{th}$ normal mode was given by
$$
|e_n\rangle = A_n\begin{bmatrix} \sin k_n x_1  \\ \sin k_n x_2  \\ \vdots \end{bmatrix}, \text{where } A_n = \sqrt{\frac{2}{N+1}}
$$
For the continous string, this eigenvector is a column vector with infinite entries, so we represent it with the eigenfunction
$$
y_n(x) = A_n \sin (k_n x) = \sqrt{\frac{2}{N+1}} \sin\left( \frac{n \pi x}{L}\right) = \sqrt{\frac{2}{L}} \sin\left( \frac{n \pi x}{L}\right)
$$
We can apply the inner product definition to this continous system. The eigenfunction and initial position column vector both have infinite entries corresponding to infinite objects, so we can express the infinite sum of their inner product as an integral:
$$
\langle y_n | y_0 \rangle = \begin{bmatrix} y_{n1}^*  & y_{n2}^* & y_{n3}^* & \cdots \end{bmatrix}\begin{bmatrix} y_{01} \\ y_{02} \\y_{02} \\ \vdots \end{bmatrix} = \sum_{j=1}^N y_{nj}^* y_{0j} 
$$
$$
\langle y_n (x) | y_0 (x) \rangle \equiv \int_0^L y_n^*(x) y_0(x) dx
$$
Similarly to the 3 beaded string, the system can be modelled as the superposition of normal modes:
$$
y(x,t) = \text{Re} \left[ \sum_{n=1}^\infty C_n e^{i\omega_n t}y_n(x)\right] \text{ , where } \text{Re}C_n = \langle y_n(x) | y_0(x) \rangle \text{ and } \text{Im}C_n = -\frac{1}{\omega_n}\langle y_n(x) | \dot{y_0}(x) \rangle
$$
Since the system begins at rest, $\dot{y_0}$ is the zero vector. The dot product of any vector with the zero vector is zero, so $\text{Im}C_n = 0$, and
$$
A_n  = \sqrt{(ReC_n)^2 + (ImC_n)^2} = \sqrt{(ReC_n)^2 + 0} = ReC_n = \langle y_n(x) | y_0(x) \rangle = \int_0^L y_n^*(x) y_0(x) dx
$$
$$
\phi_n = \tan^{-1} \left( \frac{\text{Im}C_n}{\text{Re}C_n}\right) = \tan^{-1} \left( \frac{0}{\text{Re}C_n}\right) = 0
$$
Substituting, 
$$
y(x,t) = \text{Re} \left[ \sum_{n=1}^\infty C_n e^{i\omega_n t}y_n(x)\right]
$$
$$
y(x,t) = \text{Re} \left[ \sum_{n=1}^\infty
\underbrace{\int_0^L y_n^*(x) y_0(x) dx}_{C_n} \cdot \underbrace{\exp\left(i \frac{n \pi}{L}\sqrt{\frac{TL}{M}}t\right)}_{e^{i\omega_n t}} \cdot \underbrace{\sqrt{\frac{2}{L}} \sin\left( \frac{n \pi x}{L}\right)}_{y_n(x)}
\right]
$$
$$
y(x,t) = \sum_{n=1}^\infty
\text{Re} \left[\int_0^L y_n^*(x) y_0(x) dx \right] \cdot \underbrace{\cos\left(\frac{n \pi}{L}\sqrt{\frac{TL}{M}}t\right)}_{\text{Re} \left[ e^{i \omega_n t} \right]} \cdot \sqrt{\frac{2}{L}} \sin\left( \frac{n \pi x}{L}\right)
$$
Taking advantage of initial conditions to set bounds of integration,
$$
\int_0^L y_n^*(x) y_0(x) dx = \int_{\frac{L}{2}}^{\frac{3L}{4}} \sqrt{\frac{2}{L}} \sin\left( \frac{n \pi x}{L}\right) \cdot B dx = 
\frac{B\sqrt{2L}}{\pi n} \left[ \cos\left( \frac{\pi n }{2} \right) - \cos\left( \frac{3 \pi n }{4} \right)\right]
$$

$$
y(x,t) = \sum_{n=1}^\infty
\underbrace{\frac{B\sqrt{2L}}{\pi n} \left[ \cos\left( \frac{\pi n }{2} \right) - \cos\left( \frac{3 \pi n }{4} \right)\right]}_{\text{evaluated integral}}  \cdot \cos\left(\frac{n \pi}{L}\sqrt{\frac{TL}{M}}t\right) \cdot \sqrt{\frac{2}{L}} \sin\left( \frac{n \pi x}{L}\right)
$$

$$
y(x,t) = \sum_{n=1}^\infty
\underbrace{\frac{2B}{\pi n} \left[ \cos\left( \frac{\pi n }{2} \right) - \cos\left( \frac{3 \pi n }{4} \right)\right]}_{\text{amplitude of mode $n$}} \cdot \underbrace{\cos\left(\frac{n \pi t}{L}\sqrt{\frac{TL}{M}}\right)}_{\text{temporal variation}} \cdot  \underbrace{\sin\left( \frac{n \pi x}{L}\right)}_{\text{spatial variation}}
$$
