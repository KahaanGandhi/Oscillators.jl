### Maxwell's Equations

Four equations form the foundation for electromagnetic phenomena, collectively referred to as Maxwell's equations: 
$$
\begin{align*}
& \textbf{Differential Form} & \quad & \textbf{Integral Form} \\ 
\text{Gauss's Law} & \quad \nabla \cdot \vec{E} = \frac{\rho}{\epsilon_0} & \quad & \oint \vec{E} \cdot d\vec{a} = \frac{Q_{\text{enc}}}{\epsilon_0}  \\
\text{Gauss's Law for Magnetism} & \quad \nabla \cdot \vec{B} = 0 & \quad & \oint \vec{B} \cdot d\vec{a} = 0 \\
\text{Faraday's Law of Induction} & \quad \nabla \times \vec{E} = -\frac{\partial \vec{B}}{\partial t} & \quad & \oint \vec{E} \cdot d\vec{l} = -\frac{d\Phi_B}{dt} \\
\text{Amp√®re's Law} & \quad \nabla \times \vec{B} = \mu_0 \vec{J} + \mu_0 \epsilon_0 \frac{\partial \vec{E}}{\partial t} & \quad & \oint \vec{B} \cdot d\vec{l} = \mu_0 I_{\text{enc}} + \mu_0 \epsilon_0 \frac{d\Phi_E}{dt}
\end{align*}
$$
**Gauss's law**: the integral form states that the electric field due to a point charge is given by $E = \frac{1}{4\pi \epsilon_0}\frac{Q}{r^2}$, and that the electric field flux passing through any closed surface is proportional to the enclosed charge. The differential form states that electric charge diverges from positive charge and converges upon negative charge. 

**Gauss's law for Magnetism**: the integral form states that the total magnetic flux passing through any closed surface as 0. There are no magnetic monopoles, and magnetic flux lines always close upon themselves. The differential form states that the divergence of the magnetic field at any point is zero. 

**Faraday's Law**: the integral form, also given by $\oint \vec{E} \cdot d\vec{l} = -\int \frac{\partial \vec{B}}{\partial t} \cdot d\vec{a}$, states that the changing magnetic flux through a surface induces an electromototive force (EMF), so a changing magnetic field induces a circulating electric field. The voltage of the closed circuit is proportional to the time rate of change of magnetic flux. The differential form states that a circulating magnetic field is produced by a time-varying magnetic field. 

**Ampere's Law**: the integral form states that an electric field $I$ or changing electric flux through a surface produces a circulating magnetic field. Electric currents and time-variation of $E$ are proportional to the magnetic fields. The differential form states that a circulating magnetic field is produced by an electric current and by a time-varying electric field. 

### Plane Electromagnetic Waves

Recall that $\hat{i}$ is a unit vector in the x-direction, $\hat{j}$ is a unit vector in the y-direction, and $\hat{k}$ is a unit vector in the z-direction. 

Consider electromagnetic waves propogating in the $\hat{i}$ direction, with the electric field $\vec{E}$ pointing in the $\hat{j}$ direction and magnetic field $\vec{B}$ pointing in the $\hat{k}$ direction. That is, 
$$
E = E(x,t) \hat{j} \quad \text{and} \quad B = B(x,t) \hat{k}
$$
At any time, $E$ and $B$ are uniform across the plane perpendicular to the direction of propogation, making this a plane wave. We will use Maxwell's equations to find a relationship between the strenght of the electric and magnetic fields. Consider a rectangular loop on the $xy$ plane (parellel to $\hat{j}$, the direction of the electric field). The loop goes from $x$ to $x + \Delta x$ and from $y$ to $y + \Delta y$. 

By Faraday's law, 
$$
\oint \vec{E} \cdot d\vec{s} = - \frac{d}{dt}\int \int \vec{B} \cdot d\vec{A}
$$
Using the following approximation (which becomes exact as $a \rightarrow 0$), we can first evaluate the LHS:
$$
E (x + \Delta x) \approx E(x) + \frac{\partial E}{\partial x}\Delta x
$$
$$
\oint \vec{E} \cdot d\vec{s} = \underbrace{\int_{\text{bottom left}}^{\text{bottom right}}\vec{E} \cdot d\vec{l}}_{\text{perpendicular}\rightarrow 0} + \underbrace{\int_{\text{bottom right}}^{\text{top right}}\vec{E} \cdot d\vec{l}}_{\text{parellel}\rightarrow 1} + \underbrace{\int_{\text{top right}}^{\text{top left}}\vec{E} \cdot d\vec{l}}_{\text{perpendicular}\rightarrow 0} + \underbrace{\int_{\text{top left}}^{\text{bottom left}}\vec{E} \cdot d\vec{l}}_{\text{anti-parellel}\rightarrow -1}
$$
$$
\oint \vec{E} \cdot d\vec{s} = E(x + \Delta x)\Delta y - E(x)\Delta y
$$
$$
\oint \vec{E} \cdot d\vec{s} = \left(E(x) + \frac{\partial E}{\partial x}\Delta x \right) \Delta y - E(x)\Delta y
$$
$$
\oint \vec{E} \cdot d\vec{s} = \frac{\partial E}{\partial x}\Delta x \Delta y
$$
To evaluate the RHS, we first use the right-hand rule to see that the magnetic field vector always points in the same direction as the area vector (normal to the surface of the loop). Therefore,
$$
-\frac{d}{dt}\int \int \vec{B} \cdot d\vec{A} = - \frac{\partial B}{\partial t} \Delta x \Delta y
$$
Putting together the evaluated forms of the LHS and RHS, 
$$
\frac{\partial E}{\partial x}\Delta x \Delta y = - \frac{\partial B}{\partial t} \Delta x \Delta y
$$
$$\boxed{ \boxed{
\frac{\partial E}{\partial x} = - \frac{\partial B}{\partial t}}}
$$

For $E$ and $B$ to be allowed by Faraday's laws, they must satisfy the above relation. We now apply Ampere's law: 
$$
\oint \vec{B} \cdot d\vec{l} = \underbrace{\mu_0 I_{\text{enc}}}_{\text{0 in vacuum}} + \mu_0 \epsilon_0 \frac{d\Phi_E}{dt}
$$
$$
\oint \vec{B} \cdot d\vec{l} = \mu_0 \epsilon_0 \frac{d\Phi_E}{dt}
$$
Consider a similar rectangular loop in the $xz$ plane (parellel to $\hat{k}$), the direction of the magnetic field). Taking LHS line integral, 
$$
\oint \vec{B} \cdot d\vec{s} = B(x) \Delta z - B(x + \Delta x)\Delta z
$$
$$
\oint \vec{B} \cdot d\vec{s} = B(x) \Delta z - \left(B(x) + \frac{\partial B}{\partial x}\Delta x\right)\Delta z
$$
$$
\oint \vec{B} \cdot d\vec{s} = -\frac{\partial B}{\partial x}\Delta x \Delta z
$$
Now, taking the time-derivative of electric flux on the RHS, 
$$
\mu_0 \epsilon_0 \frac{d\Phi_E}{dt} = \mu_0 \epsilon_0 \frac{d}{dt} \int \int \vec{E}\cdot d\vec{A}
$$
$$
\mu_0 \epsilon_0 \frac{d\Phi_E}{dt} = \mu_0 \epsilon_0 \frac{\partial E}{\partial t}\Delta x \Delta z
$$
Putting together the evaluated forms of the LHS and RHS, 
$$
-\frac{\partial B}{\partial x}\Delta x \Delta z = \mu_0 \epsilon_0 \frac{\partial E}{\partial t}\Delta x \Delta z
$$
$$\boxed{ \boxed{
\frac{\partial B}{\partial x} = - \mu_0 \epsilon_0 \frac{\partial E}{\partial t}}}
$$

Therefore, time-varying electric fields are generated by spatially-varying magnetic fields. We will now verify that both electric and magnetic fields satisfy the wave equation. Taking the partial derivative of $\frac{\partial E}{\partial x} = - \frac{\partial B}{\partial t}$ with respect to $x$, 
$$
\frac{\partial^2 E}{\partial x^2} = \underbrace{- \frac{\partial}{\partial x}\left(\frac{\partial B}{\partial t}\right) = - \frac{\partial}{\partial t}\left(\frac{\partial B}{\partial x}\right)}_{\text{interchanging partial derivatives}} = -\frac{\partial}{\partial t}\left(- \mu_0 \epsilon_0 \frac{\partial E}{\partial t}\right) = \mu_0 \epsilon_0 \frac{\partial^2 E}{\partial t^2}
$$
Taking the partial derivative of $\frac{\partial B}{\partial x} = - \mu_0 \epsilon_0 \frac{\partial E}{\partial t}$ with respect to $x$, 
$$
\frac{\partial^2 B}{\partial x^2} = -\frac{\partial}{\partial x} \left(\mu_0 \epsilon_0 \frac{\partial E}{\partial t}\right) = -\mu_0 \epsilon_0  \frac{\partial}{\partial t} \left(\frac{\partial E}{\partial x}\right) = -\mu_0 \epsilon_0  \frac{\partial}{\partial t} \left( - \frac{\partial B}{\partial t}\right) = \mu_0 \epsilon_0   \frac{\partial^2 B}{\partial t^2}
$$
We can express these relations as 
$$
\left(\frac{\partial^2}{\partial x^2}  - \mu_0 \epsilon_0 \frac{\partial^2 }{\partial t^2}\right) \begin{cases}\begin{rcases}E(x,t)\\ B(x,t) \end{rcases}\end{cases} = 0
$$
The general form for the one-dimensional wave equation, for wavespeed $v$ and wavefunction $\Psi$ is given by 
$$
\left(\frac{\partial^2}{\partial x^2}  - \frac{1}{v^2}\frac{\partial^2 }{\partial t^2}\right) \Psi(x,t)
$$
Therefore, $E$ and $B$ satisfy the equation and propogate at the speed of light:
$$
v =\frac{1}{\sqrt{\mu_0 \epsilon_0}} = 2.998 \times 10^8 \: \text{m/s} = c
$$
Therefore, light is an electromagnetic wave. The ratio of the amplitudes of their fields is 
$$
\frac{E}{B} = \frac{E_0}{B_0} = \frac{\omega}{k} = c
$$
The propogation direction is given by the Poynting vector 
$$
S = \frac{1}{\mu} E \times B
$$
