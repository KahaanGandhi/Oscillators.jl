### Waves on Transmission Lines

We can apply our understanding of electromagnetic waves (`Maxwell'sEquations.ipynb`) to the transmission of information. For transmission line problems, the long wavelength limit of circuit theory approximates the exact solutions given by Maxwell's equations. In essence, all circuit problems are electromagnetism problems, and can be solved through the following relations: 

$$
V = -\int\vec{E}\cdot d\vec{l} \quad \text{and} \quad I = -\int \vec{B} \cdot d\vec{l}
$$

Information can be conveyed using either the time-varying voltage between two conductors  or the time-varying current flowing through them. In the latter case, the current enters the target device through one wire and exits through the other.

**Transmission Line Model**: we think of transmission lines like pipes to guide EM waves, with capacitance and inductance distributed across its lenght. In the lumped element model (an approximation), we imagine the line as a series of very small, repeating units. Each units contains an inductor (storing magnetic energy) and a capacitor (storing electric energy). Inductors are isomorphic to beads on a string, and capacitors bridge two condcutors at each small unit. This model brekas the line into smaller segments to break down the travelling of voltage and current waves. 

**Deriving the Telegrapher's Equations**: consider the small segment of the line of lenght $\Delta x$, with inductance $L\Delta x$ and capacitance $C\Delta x$. The voltage drop of this inductor is therefore given by 
$$
V = L\frac{dI}{dt} \Longrightarrow \Delta V = - L\Delta x \frac{dI}{dt}
$$
Note the negative sign, coming from the calculation $V - (V + \Delta V)$.  Rearranging and taking the limit as $\Delta x \rightarrow 0$, we get the first Telegrapher's equation:
$$
\lim_{\Delta x \rightarrow 0} \left[ 
    \frac{\Delta V}{\Delta x} = - L \frac{dI}{dt}
\right] \Longrightarrow \boxed{\boxed{\frac{\partial V}{\partial x} = - L \frac{\partial I}{\partial t}}}
$$
Following a similar proces for capacitance, we get the second Telegrapher's equation:
$$
I = C\frac{dV}{dt} \Longrightarrow \Delta I = - C\Delta x \frac{dV}{dt}
$$
$$
\lim_{\Delta x \rightarrow 0} \left[ 
    \frac{\Delta I}{\Delta x} = - C \frac{dV}{dt}
\right] \Longrightarrow \boxed{\boxed{\frac{\partial I}{\partial x} = - C \frac{\partial V}{\partial t}}}
$$

**Deriving Wave Equations from Telegrapher's Equations**: we will differentiate each of the above equations and substitute terms from the other equation. 

$$
- L \frac{\partial^2 I}{\partial t^2} = \frac{\partial}{\partial t}\frac{\partial V}{\partial x}= \frac{\partial}{\partial x}\frac{\partial V}{\partial t} = \frac{\partial}{\partial x} \left( -\frac{1}{C}\frac{\partial I}{\partial x}\right) =   -\frac{1}{C}\frac{\partial^2 I}{\partial x^2}
$$
$$
\frac{\partial^2 I}{\partial x^2} = LC \frac{\partial^2 I}{\partial t^2}
$$
$$
-C \frac{\partial^2 V}{\partial t^2} = \frac{\partial}{\partial t}\frac{\partial I}{\partial x} = \frac{\partial}{\partial x}\frac{\partial I}{\partial t} = \frac{\partial}{\partial x}\left( -\frac{1}{L}\frac{\partial V}{\partial x}\right) =  -\frac{1}{L}\frac{\partial^2 V}{\partial x^2}
$$
$$
\frac{\partial^2 V}{\partial x^2} = LC \frac{\partial^2 V}{\partial t^2}
$$
We can express these relations as 
$$
\left(\frac{\partial^2}{\partial x^2}  - LC \frac{\partial^2 }{\partial t^2}\right) \begin{cases}\begin{rcases}V(x,t) \\ I(x,t) \end{rcases}\end{cases} = 0
$$
The general form for the one-dimensional wave equation, for wavespeed $v$ and wavefunction $\Psi$ is given by 
$$
\left(\frac{\partial^2}{\partial x^2}  - \frac{1}{v^2}\frac{\partial^2 }{\partial t^2}\right) \Psi(x,t)
$$
Therefore, $V$ and $I$ satisfy the wave equation, and have the propogation speed
$$\boxed{\boxed{
v = \frac{1}{\sqrt{LC}}}}
$$
Knowing that solutions to the wave equation are usually of the form 
$$
V(x,t) = V_0 e^{i(kx - \omega t)} \quad \text{and} \quad I(x,t) = I_0 e^{i(kx - \omega t)} 
$$
Substituing into the first Telegrapher's equation, we find the characteristic impedance of the transmission line: 
$$
\frac{\partial}{\partial x} V_0 e^{i(kx - \omega t)} = - L \frac{\partial}{\partial t} I_0 e^{i(kx - \omega t)} 
$$
$$
ik V_0 e^{i(kx - \omega t)} =  iL\omega  I_0 e^{i(kx - \omega t)} 
$$
$$
k V_0  =  L\omega  I_0 
$$
$$
\frac{\omega}{v} V_0 = L \omega I_0 \Longrightarrow \sqrt{LC} V_0 = L I_0 \Longrightarrow \frac{V_0}{I_0} = \frac{L}{\sqrt{LC}}
$$
$$\boxed{\boxed{
Z_0 = \sqrt{\frac{L}{C}}}}
$$

**Left and Right Moving Waves**: 

$$
\underbrace{V(x,t) = Z_0 I(x,t)}_\text{right moving} \quad \text{and} \quad \underbrace{V(x,t) = -  Z_0 I(x,t)}_\text{left moving}
$$
For right moving waves (travelling in the +$x$-direction), current flowing to the right on the top wire is in phase with the voltage, and current at any given time on the bottom wire is always the opposite of the top wire.
