### Power of a Sinusoidal Rope Wave

To find the power of a sinusoidal rope wave, first consider the kinetic energy (KE) of a small rope element $dm$. 

$$
KE = \frac{1}{2} dm\cdot v^2, \text{ where } dm = \mu dx
$$
$$
y(x,t) = A \sin (kx - \omega t) \Longrightarrow v(x,t) = -\omega  A \cos (kx - \omega t)
$$
$$
KE = \frac{1}{2} \mu dx \cdot \omega^2  A^2 \cos^2 (kx - \omega t)
$$

We can integrate over a full wavelength to find total KE remembering that the average value of $\cos^2$ over a cycle is $\frac{1}{2}$, 

$$
KE_{total} = \int_0^\lambda  \frac{1}{2} \mu  \omega^2  A^2 \cos^2 (kx - \omega t) dx = \frac{1}{4} \mu  \omega^2  A^2 \lambda
$$

By following the same procedure for potential energy (U), we arrive a similar result: 

$$
U_{total} = \frac{1}{4} \mu  \omega^2  A^2 \lambda
$$

A period of a cycle is $\frac{\lambda}{v}$. Since power is the rate of energy transfer, 
$$
P = \frac{E}{t} = \frac{KE_{total} + U_{total}}{\frac{\lambda}{v}} = \frac{1}{2} \mu  \omega^2  A^2 v
$$
Note that $P \propto A^2$. 

### Power and Intensity of EM Waves
The energy density of E and B are given by 
$$
\eta_E = \frac{\epsilon_0}{2}E^2 \quad \text{and} \quad \eta_B = \frac{1}{2\mu_0}B^2
$$
As we showed in `Maxwell'sEquations.ipynb`, $E = cB$ for EM waves in a vacuum. Substituting,
$$
\eta_B = \frac{1}{2\mu_0}\frac{E^2}{c^2} = \frac{1}{2\mu_0}\frac{E^2}{\frac{1}{\mu_0 \epsilon_0}} = \frac{\epsilon_0}{2}E^2
$$
Therefore, the total energy density of EM radiation is
$$
\eta_{rad} = \eta_E + \eta_B = \frac{\epsilon_0}{2}E^2 + \frac{\epsilon_0}{2}E^2 = \epsilon_0 E^2
$$
For total energy, consider a box with volume $A \times dx$ moving wih the wave. The  energy enclosed by this box is 
$$
\eta_{rad} \times Adx = \epsilon_0 E^2 A dx
$$
As the box moves with the wave, the energy in the box is transfered in time $dt = \frac{dx}{c}$, so power and intensity ($S$) are given by 
$$
P = \frac{E}{t} = \frac{ \epsilon_0 E^2 A dx}{\frac{dx}{c}} = c\epsilon_0 E^2 A 
$$
$$
S = \frac{P}{A} = c\epsilon_0 E^2
$$
Intensity, power per unit area, can also be represented using $E = cB$: 
$$
S = c\epsilon_0 E(cB) = c^2 \epsilon_0 EB = \left( \frac{1}{\mu_0 \epsilon_0 }\right) \epsilon_0 EB = \frac{1}{\mu_0} EB
$$
Note that the Poynting vector, which points in the direction of propogation, has magnitude equal to the intensity of the wave:
$$
S = \frac{1}{\mu_0}E\times B
$$
With a formula for instantenous intensity, we can now define average intensity (with angle brackets indicated the average over one period): 
$$
\langle S \rangle = c \epsilon_0 \langle E^2 \rangle
$$
Defining root mean square (rms, the square root of the mean of the square) amplitude in terms of this average, 
$$
E_{rms} = \sqrt{\langle E^2 \rangle}
$$
Applying this new definition, 
$$
\langle S \rangle = c \epsilon_0 E_{rms}^2
$$
$$
E = cB \Rightarrow E_{rms} = cB_{rms}
$$
$$
\langle S \rangle = c^2 \epsilon_0  E_{rms} B_{rms} = \frac{1}{\mu_0} E_{rms} B_{rms}
$$
Therefore, we see that the Poynting vector can represent both instantanous and average cases. 

Note that for a sinusoid, 
$$
\langle E^2 \rangle = \frac{E^2_{peak}}{2}
$$
$$
E_{rms} \equiv \sqrt{\langle E^2 \rangle} = \sqrt{\frac{E^2_{peak}}{2}} = \frac{E_{peak}}{\sqrt{2}}
$$
